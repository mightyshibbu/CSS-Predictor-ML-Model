```python
import tinycss2
```

```python
with open('Auction.css', 'r') as css_file:
    css_content = css_file.read()
rules = tinycss2.parse_stylesheet(css_content, skip_comments=True, skip_whitespace=True)
```

```python
print(rules[0])
```

```python

# Iterate through the rules and filter out irrelevant tokens
for rule in rules:
    if rule.type == 'qualified-rule':  # This ensures we're dealing with CSS rules
        selector = ''.join([token.serialize() for token in rule.prelude])  # Convert tokens to readable text
        print(f'Selector: {selector}')
            
        # Parse the declarations within the rule
        declarations = tinycss2.parse_blocks_contents(rule.content)
        for declaration in declarations:
            if declaration.type == 'declaration':  # Only process declaration tokens
                print(f'{declaration.name}: {declaration.value}')
```
